# Drop Columns

```python
#drop unnecessary columns
df_1 = df.drop(columns = {'CAMIS', 'BUILDING', 'STREET', 'PHONE', 'Community Board', 'Council District', 'Census Tract', 'BIN',
                  'BBL', 'NTA', 'Location Point1', 'Zip Codes', 'Community Districts', 'Borough Boundaries',
                  'City Council Districts', 'Police Precincts'})
```

```python
df_1
```

# Rename Columns

```python
#rename DBA column to NAME
df_1.rename(columns = {'DBA':'NAME'}, inplace = True)
```

```python
# change name of CANTINA CAQUERIA & TEQUILIA BAR to CANTINA TAQUERIA & TEQUILIA BAR
df_1.rename({'CANTINA CAQUERIA & TEQUILIA BAR' : 'CANTINA TAQUERIA & TEQUILIA BAR'}, inplace = True)
```

# Data Types

```python
#check data types
df_1.dtypes
```

```python
#check for mixed types
for col in df_1.columns.tolist():
  weird = (df_1[[col]].map(type) != df_1[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_1[weird]) > 0:
    print (col)
```

```python
#change to uniform data types
df_1['VIOLATION CODE'] = df_1['VIOLATION CODE'].astype('str')
```

```python
df_1['VIOLATION DESCRIPTION'] = df_1['VIOLATION DESCRIPTION'].astype('str')
```

```python
df_1['GRADE'] = df_1['GRADE'].astype('str')
```

```python
#change data type to date
df_1['GRADE DATE'] = pd.to_datetime(df_1['GRADE DATE'])
```
